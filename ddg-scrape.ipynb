```python
# Web Scraping
from bs4 import BeautifulSoup
import requests

# Data manipulation
import pandas as pd
import numpy as np
import json
from datetime import datetime 
```

```python
# will use appended data if there is an error and run different scrapes
appended_data = []
```

```python
# day and month for ease of code 
day = 21
month = 7
month_malay = 'julai'
```

```python
# will use the below script if want to run individually
```

```python
# this code is to run a loop for multiple days / pages 
```

```python
tarikh_col =[]
nama_kluster = []
kategori = []
negeri =[]
daerah = []
kes = []
saring = []  


for i in range(21,23):
    
    # Go to first page of the money section
    url_main = "https://kpkesihatan.com/2021/07/{}/kenyataan-akhbar-kpk-{}-julai-2021-situasi-semasa-jangkitan-penyakit-coronavirus-2019-covid-19-di-malaysia/".format(i,i)

    r = requests.get(url_main)
    soup = BeautifulSoup(r.content, 'html.parser')
    data = soup.findAll('table')
    
    tarikh = datetime(2021,7,i)
    print(tarikh)
    print(len(data))
    
    
    for i in data[-3].find_all('tr')[1:]:
        x = i.find_all('td')
        tarikh_col.append(tarikh)
        nama_kluster.append(x[1].text)
        kategori.append(x[2].text)
        negeri.append(x[3].text)
        daerah.append(x[4].text)
        kes.append(x[5].text)
        saring.append(x[6].text)    
```

```python

#create new df 
df = pd.DataFrame({'tarikh':tarikh_col,'nama_kluster':nama_kluster, 'kategori_kluster': kategori, 'negeri' : negeri, 'daerah': daerah, 
                   'jumlah_kes':kes, 'jumlah_disaring': saring })
# append into appended df 
appended_data.append(df)
```

```python
# concat the new data 
df_new = pd.concat(appended_data)
```
